```python
# prompt: df['molecular_weight']에서 각 데이터에 '['이 포함됐는지 아닌지 알려줘. df는 pandas dataframe 형태야

dfnum = df['molecular_weight'].map(lambda x: '[' in x)
```

```python
dfnum
```

```python
# prompt: dfnum에서 True인것을 df에서 None으로 바꾸는 코드를 만들어줘

df['molecular_weight'] = df['molecular_weight'].map(lambda x: None if '[' in x else x)
```

```python
df.info()
```

```python
# prompt: df['molecular_weight']의 Dtype을 float 형태로 바꿔줘

df['molecular_weight'] = df['molecular_weight'].map(lambda x: float(x) if x is not None else np.nan)
df.info()
```
